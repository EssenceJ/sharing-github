```python
#Essence Joe
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x = [5,8,10]
y = [12,16,6]
x2 = [6,9,11]
y2 = [6,15,7]
plt.plot(x,y,'r',label = 'line one', linewidth = 5)
plt.plot(x2,y2,'c',label = 'line two', linewidth = 5)
plt.show()
```

```python
plt.bar(x,y, label='Sample one',color='g')
plt.bar(x2,y2, label='Sample two', color='r')
plt.legend()
plt.xlabel('Bar number')
plt.ylabel('Bar height')
plt.title('Bar graph')
plt.show()
```

```python
pop_ages = [22,25,62,45,67,86,45,34,67,87,54,53,56,90,4,34,78,90,65,34,55,34,27,38,59,69,78,65,74,73,45,78,90,29,87,54,23,19,65,79,85,43,45]
bins = [0,10,20,30,40,52,60,70,80,90]
plt.hist(pop_ages, bins, histtype ='bar', rwidth = 0.8)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Histogram')
plt.legend()
plt.show()
```

```python
xs = [1,2,3,4,5,6,7,8]
ys = [5,2,4,2,1,4,5,2]
plt.scatter(xs,ys, label = 'skitscat', color = 'k')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter Plot')
plt.legend()
plt.show()
```

```python
slices = [7,2,2,13]
activities = ['sleeping','eating','working','playing']
#column colors
cols = ['c','m','r','b']
plt.pie(slices,
        labels = activities,
        colors = cols,
        startangle = True,
        shadow = True,
        #pulling out slices
        explode = (0,0.1,0,0),
        autopct = '%1.1f%%')
plt.title('Pie Plot')
plt.show()
```

```python
days = [1,2,3,4,5]

sleeping = [7,8,6,11,7]
eating = [2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color = 'm', label = 'Sleeping', linewidth = 5)
plt.plot([],[],color = 'c', label = 'Eating', linewidth = 5)
plt.plot([],[],color = 'r', label = 'Working', linewidth = 5)
plt.plot([],[],color = 'k', label = 'Playing', linewidth = 5)

plt.stackplot(days,sleeping,eating,working,playing, colors = ['m','c','r','k'])

plt.xlabel('x')
plt.ylabel('y')
plt.title('Area Plot')
plt.legend()
plt.show()
```

```python
import numpy as np
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
t1 = np.arange(0.0,5.0,0.1)
t2 = np.arange(0.0,5.0,0.02)

plt.subplot(211)
plt.plot(t1,f(t1),'bo',t2,f(t2))

plt.subplot(212)
plt.plot(t2,np.cos(2*np.pi*t2))
plt.show()
```
